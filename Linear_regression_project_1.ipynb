```python
### Feature engineering
```

```python
### Label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
```

```python
data_1['Gender_Encode'] = Label_Encoder.fit_transform(data_1['Gender'])
```

```python
data_1['Degree_Encode'] = Label_Encoder.fit_transform(data_1['Degree'])
```

```python
data_1['Job_Title_Encode'] = Label_Encoder.fit_transform(data_1['Job_Title'])
```

```python
### Data after label encoding 
```

```python
data_1.head()
```

```python
### Feature Scalling
```

```python
### import standard scaler
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
data_1['Age_Scaled'] = std_scaler.fit_transform(data_1[['Age']])
data_1['Experience_years_Scaled'] = std_scaler.fit_transform(data_1[['Experience_years']])
```

```python
### Feature after scalling 
```

```python
data_1.head()
```

```python
### Dependent and independent features
```

```python
x = data_1[['Age_Scaled', 'Gender_Encode', 'Degree_Encode', 'Job_Title_Encode', 'Experience_years_Scaled']]
y = data_1['Salary']
```

```python
### Dependent features 
```

```python
x.head()
```

```python
### Splitting training and testing features 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.head()
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
y_test.head()
```

```python
### Model developement 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Linear_regression_model = LinearRegression()
```

```python
### Model training 
```

```python
Linear_regression_model.fit(x_train, y_train)
```

```python
### Model prediction 
```

```python
y_pred_lr = Linear_regression_model.predict(x_test)
y_pred_lr
```

```python
### Create dataframe
```

```python
df = pd.DataFrame({'y_Actual' : y_test, 'y_Predicted' : y_pred_lr })
df
```

```python
df = pd.DataFrame({'y_Actual' : y_test, 'y_Predicted' : y_pred_lr })
df['Error'] = df['y_Actual'] - df['y_Predicted']
df
```

```python
df = pd.DataFrame({'y_Actual' : y_test, 'y_Predicted' : y_pred_lr })
df['Error'] = df['y_Actual'] - df['y_Predicted']
df['abs_error'] = abs(df['Error'])
df
```

```python
Mean_absolute_Error = df['abs_error'].mean()
Mean_absolute_Error
```

```python
### Model evaluation 
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
### Model accuracy
```

```python
r2_score(y_test, y_pred_lr)
```

```python
print(f'Accuracy of the model = {round(r2_score(y_test, y_pred_lr), 4)*100}%')
```

```python
### Mean Absolute Error
```

```python
round(mean_absolute_error(y_test, y_pred_lr),2)
```

```python
print(f"Mean Absolute Error  = {round(mean_absolute_error(y_test, y_pred_lr),2)}")
```

```python
### Mean Squared Error
```

```python
Mean_square_error  = round(mean_squared_error(y_test, y_pred_lr),2)
Mean_square_error
```

```python
print(f'Mean Squared Error = {round(mean_squared_error(y_test, y_pred_lr),2)}')
```

```python
### Root Mean Squared Error
```

```python
print('Root Mean Squared Error (RMSE) = ', Mean_square_error**(0.5))
```
